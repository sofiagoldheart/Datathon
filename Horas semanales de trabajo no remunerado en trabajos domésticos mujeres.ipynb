```python
# Librerias utilizadas para la representación y análisis de datos
import pandas as pd
import matplotlib.pyplot as plt
```

# HORAS PROMEDIO SEMANALES DEDICADAS A REALIZAR TRABAJO DOMÉSTICO NO REMUNERADO PARA EL HOGAR Y PARA OTROS HOGARES

```python
# Load the dataset
datathonset = pd.read_csv('/content/drive/MyDrive/DatasetsDatathon/horas_trabajo_no_remunerado.csv', encoding='ISO-8859-1')
# Visualize the dataset
datathonset
```

```python
print(datathonset.columns)
```

```python
columns_to_visualize = ['hrs_promedio_total', 'hrs_promedio_hombres', 'hrs_promedio_mujeres']

# Create a bar chart for each column
for column in columns_to_visualize:
    plt.figure(figsize=(10, 6))  # Set the size of the figure
    plt.bar(datathonset['Entidad'], datathonset[column])  # Create the bar chart
    plt.xlabel('Entidad')  # X-axis label
    plt.ylabel('Hours')  # Y-axis label
    plt.title(f'Average Hours Worked - {column}')  # Chart title
    plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
    plt.show()  # Display the chart
```
